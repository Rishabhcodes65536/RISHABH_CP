```python
def lagrange(x, y, xi):
    n = len(x)
    result = 0.0

    for i in range(n):
        res = y[i]
        for j in range(n):
            if i != j:
                res *= (xi - x[j]) / (x[i] - x[j])
        result += res

    return result

# Given data points
x = [0, 1, 2]
y = [1, 3, 2]

# Points where you want to compute the polynomial values
x1 = 0.5
x2 = 1.5

# Calculate polynomial values
result1 = lagrange(x, y, x1)
result2 = lagrange(x, y, x2)

print(f"At x={x1}, the interpolated value is: {result1}")
print(f"At x={x2}, the interpolated value is: {result2}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def lagrange_cardinal(x, xi, i):
    result = 1.0
    for j in range(len(x)):
        if j != i:
            result *= (xi - x[j]) / (x[i] - x[j])
    return result

# Given data points
x = [0, 1, 2]
y = [1, 3, 2]

# Generate points for plotting
x_values = np.linspace(0, 2, 100)
y_values = [lagrange(x, y, xi) for xi in x_values]

# Calculate and plot cardinal functions
plt.figure(figsize=(20, 9))
for i in range(len(x)):
    y_cardinal = [lagrange_cardinal(x, xi, i) for xi in x_values]
    plt.plot(x_values, y_cardinal, label=f'C{i}(x)')

# Plot Lagrange's polynomial
plt.plot(x_values, y_values, label="Lagrange's Polynomial", linestyle='--')

# Plot data points
plt.scatter(x, y, color='red', label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cardinal Functions and Lagrange\'s Polynomial')
plt.legend()
plt.grid(True)
plt.show()
```
